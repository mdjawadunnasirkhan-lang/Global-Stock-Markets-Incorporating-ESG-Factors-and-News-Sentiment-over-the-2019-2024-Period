## Prepare Data for LSTM

### Subtask:
Preprocess the `final_merged_df` for LSTM modeling, including scaling numerical features, creating time-series sequences (e.g., look-back periods for X and corresponding y values), and splitting the data into training and testing sets.


**Reasoning**:
The first instruction is to select relevant numerical features and scale them using `MinMaxScaler`. I will extract the specified columns ('Close', 'High', 'Low', 'Open', 'Volume', 'ESG_Score', 'Title_Sentiment') into a new DataFrame and then apply `MinMaxScaler` to normalize these features.



```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# 1. Select the relevant numerical features
features = ['Close', 'High', 'Low', 'Open', 'Volume', 'ESG_Score', 'Title_Sentiment']
data = final_merged_df[features].values

# 2. Scale the selected numerical features
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

print("Selected features:", features)
print("Shape of original data:", data.shape)
print("Shape of scaled data:", scaled_data.shape)
print("First 5 rows of scaled data:\n", scaled_data[:5])
```

**Reasoning**:
Now that the numerical features are scaled, I will define the `look_back` period and then create the input sequences (X) and target values (y) for the LSTM model. These will be converted to NumPy arrays and reshaped appropriately for LSTM input.



```python
look_back = 60 # Define the look-back period (e.g., 60 days)

# Create empty lists for X and y
X, y = [], []

# Loop through the scaled data to create sequences
for i in range(look_back, len(scaled_data)):
    X.append(scaled_data[i-look_back:i])
    # The target is the 'Close' price, which is the first feature (index 0) in our 'features' list
    y.append(scaled_data[i, 0])

# Convert the lists to NumPy arrays
X = np.array(X)
y = np.array(y)

print(f"Look-back period: {look_back}")
print(f"Shape of X (samples, time_steps, features): {X.shape}")
print(f"Shape of y (samples, target_value): {y.shape}")
print("First 5 X samples:\n", X[:5])
print("First 5 y samples:\n", y[:5])
```

**Reasoning**:
Now that the input sequences (X) and target values (y) are prepared, I will split them into training and testing sets. It's crucial to maintain chronological order for time-series data, so I'll define a split point that allocates a certain percentage (e.g., 80%) for training and the remaining for testing.



```python
train_size = int(len(X) * 0.8)
X_train, X_test = X[0:train_size], X[train_size:len(X)]
y_train, y_test = y[0:train_size], y[train_size:len(y)]

print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

## Define and Train LSTM Model

### Subtask:
Define the architecture of the LSTM model using Keras/TensorFlow, compile it, and train it on the prepared training data.


**Reasoning**:
To define the LSTM model architecture, I need to import the `Sequential`, `LSTM`, and `Dense` layers from TensorFlow/Keras. I will then initialize a Sequential model and add the LSTM and Dense layers as specified in the instructions, ensuring the `input_shape` for the first LSTM layer is correctly set based on `X_train`'s dimensions.



```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# 2. Initialize a Sequential model
model = Sequential()

# 3. Add an LSTM layer
# input_shape should be (time_steps, features)
model.add(LSTM(units=50, return_sequences=False, input_shape=(X_train.shape[1], X_train.shape[2])))

# 4. Add a Dense output layer
model.add(Dense(units=1))

# Print the model summary to verify the architecture
model.summary()
print("LSTM model architecture defined successfully.")
```

**Reasoning**:
The previous step successfully defined the LSTM model architecture. Now, I need to compile the model by specifying an optimizer and a loss function, and then train it using the prepared training data (`X_train`, `y_train`). I will also set appropriate epochs and batch size, and include a validation split.



```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras import Input

# 2. Initialize a Sequential model
model = Sequential()

# Add an Input layer to explicitly define the input shape, addressing the UserWarning
model.add(Input(shape=(X_train.shape[1], X_train.shape[2])))

# 3. Add an LSTM layer
# input_shape should be (time_steps, features)
model.add(LSTM(units=50, return_sequences=False))

# 4. Add a Dense output layer
model.add(Dense(units=1))

# 5. Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# 6. Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.1, verbose=1)

# Print the model summary to verify the architecture
model.summary()
print("LSTM model compiled and trained successfully.")
```

## Make Predictions and Evaluate Performance

### Subtask:
Use the trained LSTM model to make predictions on the test set. Calculate and display the performance metrics: R-squared (R2), Root Mean Squared Error (RMSE), and Mean Squared Error (MSE).


**Reasoning**:
I will make predictions on the test set, inverse transform both the predictions and actual values to their original scale, and then calculate and display the R-squared, MSE, and RMSE performance metrics to evaluate the model.



```python
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

# 1. Make predictions on the test set
y_pred = model.predict(X_test)

# 2. Inverse transform both y_test and y_pred to original scale

# Create a dummy array for y_test to inverse transform it
y_test_reshaped = y_test.reshape(-1, 1)
# The scaler expects an array with the same number of features it was fitted on.
# We need to create a dummy array where the 'Close' price is in its original position (index 0)
# and other features are placeholders (e.g., zeros or means, but zeros are fine for inverse_transform if we only care about index 0).
# The number of features is scaler.n_features_in_ or len(features).

# For y_test_original:
dummy_y_test = np.zeros((len(y_test_reshaped), scaler.n_features_in_))
dummy_y_test[:,0] = y_test_reshaped[:,0] # Place the actual Close prices in the first column
y_test_original = scaler.inverse_transform(dummy_y_test)[:,0] # Inverse transform and take only the Close price

# For y_pred_original:
dummy_y_pred = np.zeros((len(y_pred), scaler.n_features_in_))
dummy_y_pred[:,0] = y_pred[:,0] # Place the predicted Close prices in the first column
y_pred_original = scaler.inverse_transform(dummy_y_pred)[:,0] # Inverse transform and take only the Close price

# 4. Calculate R-squared (R2) score
r2 = r2_score(y_test_original, y_pred_original)

# 5. Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test_original, y_pred_original)

# 6. Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# 7. Print the metrics
print(f"R-squared (R2) Score: {r2:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
```
