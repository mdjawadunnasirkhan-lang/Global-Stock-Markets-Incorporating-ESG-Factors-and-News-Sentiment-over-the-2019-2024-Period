## prepare_data_for_lstm

### Subtask:
Prepare the `final_merged_df` by selecting relevant features, scaling the data, and creating sequences for the LSTM model. This includes defining the look-back period for time series prediction.


**Reasoning**:
I will start by selecting the specified features from `final_merged_df`, define the `look_back` period, and then chronologically split the data into training and testing sets, as these are the foundational steps for preparing the data for LSTM model training.



````python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# 1. Select relevant features
features = ['Close', 'Open', 'High', 'Low', 'Volume', 'ESG_Score', 'Title_Sentiment']
data = final_merged_df[features].values

# 2. Define the look-back period
look_back = 60

# 3. Split the data into training and testing sets chronologically (80/20 split)
train_size = int(len(data) * 0.8)
train_data = data[:train_size]
test_data = data[train_size:]

print(f"Original data shape: {data.shape}")
print(f"Training data shape: {train_data.shape}")
print(f"Testing data shape: {test_data.shape}")
````

**Reasoning**:
Now that the data is split, I will scale both the training and testing sets using `MinMaxScaler` by fitting the scaler only on the training data to prevent data leakage.



````python
scaler = MinMaxScaler(feature_range=(0, 1))

# Fit on training data and transform both training and testing data
train_data_scaled = scaler.fit_transform(train_data)
test_data_scaled = scaler.transform(test_data)

print("Training data scaled successfully.")
print("Testing data scaled successfully.")
print(f"Scaled training data shape: {train_data_scaled.shape}")
print(f"Scaled testing data shape: {test_data_scaled.shape}")
````

**Reasoning**:
Now that the data has been scaled, I will define a function to create sequences for the LSTM model. This function will take the scaled data and the look-back period, generating input sequences (X) and corresponding target values (y) as specified in the instructions.



````python
def create_sequences(data, look_back):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:(i + look_back)])
        # Target value is the 'Close' price (index 0) at the end of the sequence
        y.append(data[i + look_back, 0])
    return np.array(X), np.array(y)

print("create_sequences function defined successfully.")
````

**Reasoning**:
Now that the `create_sequences` function is defined, I will apply it to both the scaled training data (`train_data_scaled`) and scaled testing data (`test_data_scaled`) using the `look_back` period to generate the input sequences (X) and target values (y) for both the training and testing sets.



````python
X_train, y_train = create_sequences(train_data_scaled, look_back)
X_test, y_test = create_sequences(test_data_scaled, look_back)

print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
````

````python
import numpy as np

# Create dummy arrays to inverse transform predictions and y_test
# The original features were ['Close', 'Open', 'High', 'Low', 'Volume', 'ESG_Score', 'Title_Sentiment']
# The target is 'Close', which is the first feature (index 0)

# For predictions: fill a zero array with predictions at the 'Close' price position
predictions_full = np.zeros((predictions.shape[0], len(features)))
predictions_full[:, 0] = predictions[:, 0]

# Inverse transform the predictions
predictions_rescaled = scaler.inverse_transform(predictions_full)[:, 0]

# For y_test: fill a zero array with y_test at the 'Close' price position
y_test_full = np.zeros((y_test.shape[0], len(features)))
y_test_full[:, 0] = y_test

# Inverse transform y_test
y_test_rescaled = scaler.inverse_transform(y_test_full)[:, 0]

print("Predictions and actual test values inverse transformed successfully.")
````
